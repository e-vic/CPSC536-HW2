$$Ax = b + w$$

$A,b$ are data, $w$ is noise. $b$ is the blurred image, the matrix $A$ describes the blur operator. We want to estimate the original (before blurring) image by finding $x$. We find $x$ by solving

$$\min_{x} ||Ax - b||_2^2 \text{ subject to } ||x||_1 \leq \tau$$

for different values of $\tau$. $A$ is a blurring matrix that we construct.
